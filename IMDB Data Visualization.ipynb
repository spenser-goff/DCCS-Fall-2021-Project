```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import seaborn as sns
```

```python
imdb_df = pd.read_csv('finished_data_wrangling.csv')
imdb_df.drop(imdb_df.columns[[0]], axis = 1, inplace = True)
sns.histplot(imdb_df, x='genre')
plt.xticks(rotation = 45)
```

```python
sns.set(rc={"figure.figsize":(16,9)})#Width = 16 Height = 9
#palette = {"Adventure":"tab:purple"}
_ = sns.histplot(imdb_df, x='genre',hue='genre')
_ = plt.xticks(rotation = 45)
_ = plt.xlabel("Genre of Movies")
_ = plt.ylabel("Number of Movies in Dataset")
_ = plt.title("Distribution of Movie Genres in the Dataset")
```

```python
_ = sns.scatterplot(data = imdb_df, x = 'IMDB_score', y =  'budget', hue= "genre")
_ = plt.xlabel("IMDB Score")
_ = plt.ylabel("Budget")
_ = plt.title("Budget for Movies across IMDB Ratings")

```

```python
_ = sns.scatterplot(data = imdb_df, x = 'IMDB_score', y =  'budget', hue = 'genre').set_yscale("log")
_ = plt.xlabel("IMDB Score")
_ = plt.ylabel("Budget")
_ = plt.title("Budget for Movies across IMDB Ratings")
```

```python
imdb_df_10s = imdb_df.loc[imdb_df['title_year'] > 2009]
graph = sns.scatterplot(data = imdb_df_10s, x = 'budget', y = 'gross', hue = 'genre', palette = 'colorblind')
graph.set(xscale="log", yscale="log")
_ = plt.xlabel("Budget")
_ = plt.ylabel("Gross")
_ = plt.title("Budget vs Gross Earnings for IMDB Movies (2010s)")
plt.show()
```

```python
_ = imdb_df.groupby('title_year').budget.min().plot(label = 'min', c = 'green').set_yscale("log")
_ = imdb_df.groupby('title_year').budget.max().plot(label = 'max', c = 'orange').set_yscale("log")
_ = plt.legend(loc='best')
_ = plt.xlabel("Year of Release")
_ = plt.ylabel("Budget")
_ = plt.title("Budget for Movies across Years (Min & Max)")
```

```python
imdb_df["net"] = imdb_df["gross"] / imdb_df["budget"]
gby = imdb_df.groupby("title_year")
gby.net.mean().plot(c = "purple", lw = 4) #lw = linewidth, c = color
_ = plt.xlabel("Year of Release")
_ = plt.ylabel("Net Profit/Loss")
_ = plt.title(" Average Net Profit/Loss for Movies across the Years")

```

```python
_ = sns.scatterplot(data = imdb_df, x = 'genre', y =  'net', hue = 'genre').set_yscale("log")
_ = plt.xlabel("Genre")
_ = plt.ylabel("Net")
_ = plt.title("Budget-Gross Ratio for Movies across Genre")

```
